```python
Cov = pd.read_csv('https://raw.githubusercontent.com/ptanaya58/Geopandas/main/Data_Harian_Kasus_per_Provinsi_COVID-19_Indonesia.csv')
Cov
```

```python
idn = geopandas.read_file("/content/gdrive/MyDrive/Tugas Pertemuan 10 Visdat/indo-polygon.shp")
idn
```

```python
idn.plot(column='Propinsi', figsize=(15, 7), legend=True)
```

```python
cv19 = idn.merge(Cov, left_on="kode", right_on="Kode_Provi")
cv19
```

```python
cv19.plot(column = 'Kasus_Posi', figsize=(15, 7), legend=True)
```
